```python
from PIL import Image
img = Image.open("images.jpeg")
```

```python
import matplotlib.pyplot as plt

plt.imshow(img)
plt.show()
```

```python

source = img.split()
R,G,B=0,1,2



# DARKNESS IMG
constant = 100.0
Red = source[R].point(lambda i:i-constant)
Green = source[G].point(lambda i:i-constant)
Blue = source[B].point(lambda i:i-constant)

# BRIGHTNESS IMG
# constant = 100.0

# Red = source[R].point(lambda i:i+constant)
# Green = source[G].point(lambda i:i+constant)
# Blue = source[B].point(lambda i:i+constant)


# INVERT IMG

# max = 255.0
# Red = source[R].point(lambda i:max-i)
# Green = source[G].point(lambda i:max-i)
# Blue = source[B].point(lambda i:max-i)




# LOWER CONTRAST IMG

# constant = 2.5

# Red = source[R].point(lambda i:i/constant)
# Green = source[G].point(lambda i:i/constant)
# Blue = source[B].point(lambda i:i/constant)

# Red = source[R].point(lambda i:i*constant)
# Green = source[G].point(lambda i:i*constant)
# Blue = source[B].point(lambda i:i*constant)




img = Image.merge(img.mode,(Red,Green,Blue))
plt.imshow(img)
plt.show()
```
